```python
from fastapi import FastAPI, HTTPException
import pandas as pd

app = FastAPI()

# Create a FastAPI endpoint to fetch all movie data
@app.get("/movies/")
async def get_movies():
    return movie.to_dict(orient="records")

# Endpoint to get top 5 movies by Rating
@app.get("/movies/top-rated/")
async def get_top_rated_movies():
    top_rated = movie.sort_values('Rating', ascending=False).head(5)
    return top_rated.to_dict(orient="records")

# Endpoint to get movies by year
@app.get("/movies/by-year/{year}")
async def get_movies_by_year(year: int):
    movies_by_year = movie[movie['Year'] == year]
    if movies_by_year.empty:
        raise HTTPException(status_code=404, detail="No movies found for this year.")
    return movies_by_year.to_dict(orient="records")

# Endpoint to get latest movies (from 2010 onward)
@app.get("/movies/latest/")
async def get_latest_movies():
    latest_movies = movie[movie['Year'] >= 2010]
    return latest_movies.to_dict(orient="records")

# Endpoint to get movies by Director
@app.get("/movies/by-director/{director}")
async def get_movies_by_director(director: str):
    movies_by_director = movie[movie['Director'].str.contains(director, case=False, na=False)]
    if movies_by_director.empty:
        raise HTTPException(status_code=404, detail="No movies found for this director.")
    return movies_by_director.to_dict(orient="records")

# Endpoint to get top 5 movies by Review Count
@app.get("/movies/top-reviewed/")
async def get_top_reviewed_movies():
    top_reviewed = movie.sort_values('Review Count', ascending=False).head(5)
    return top_reviewed.to_dict(orient="records")

# Endpoint to get movie details by Title
@app.get("/movies/by-title/{title}")
async def get_movie_by_title(title: str):
    movie_by_title = movie[movie['Title'].str.contains(title, case=False, na=False)]
    if movie_by_title.empty:
        raise HTTPException(status_code=404, detail="No movie found with this title.")
    return movie_by_title.to_dict(orient="records")


# Endpoint to get summary statistics
@app.get("/movies/stats/")
async def get_movie_statistics():
    stats = movie.describe().to_dict()
    return stats



    


```
